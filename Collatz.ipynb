```python
import matplotlib.pyplot as plt

def collatz_sequence(n):
    """Generate values in the Collatz sequence starting from n."""
    if n <= 0:
        raise ValueError("The input should be a positive integer!")
    
    sequence = [n]
    while n > 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence

def plot_collatz(n):
    """Plot the Collatz sequence for a given n, with its convergence to the 4,2,1 cycle."""
    sequence = collatz_sequence(n)
    steps = len(sequence) - 1
    plt.figure(figsize=(10, 6))
    plt.plot(sequence, 'o-', markersize=4)
    plt.yscale('log')  # Use logarithmic scale for better visualization
    plt.xlabel('Step')
    plt.ylabel('Value')
    plt.title(f'Collatz Conjecture: Convergence of {n} in {steps} steps')
    plt.grid(True, which="both", linestyle='--', linewidth=0.5)
    plt.show()
    print(f"The sequence for {n} converged in {steps} steps.")
```

```python
# Ejemplo 
a = 27
plot_collatz(a)
```
